```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "Col1": range(12),
    "Col2": ["A"]*3 + ["B"]*3 + ["C"]*3 + ["D"]*3,
    "Date":pd.to_datetime(["2026-10-11","2026-04-12","2026-09-22","2026-06-08"]*3)
})

print(df,"\n")
pivoted = df.pivot(index = "Date",columns="Col2", values="Col1")
print(pivoted)
```

```python
data = {
    'Course':['cs','civil','cs','civil','cs'],
    'year':[1,2,1,1,2],
    'Student':[100,150,120,90,180]
}

df = pd.DataFrame(data)
print("Original dataset:\n")
print(df)

pivot_data = df.pivot_table(
    values='Student',
    index='Course',
    columns='year',
    aggfunc='sum'
)

print("the pivoted table for the dataset is :\n")
print(pivot_data)
```

```python
data =[
        [1,2,3],
        [5,6,7],
        [8,9,10]
    ]

df = pd.DataFrame(data,index=['a','c','d'], columns = ['one','two','three'])
df = df.reindex(['a','b','c','d','e','f'])
res = df.ffill(limit=1)
res2 = df.bfill()
print(df,'\n')
print(res,'\n')
print(res2,'\n')
```
