```python
# 1. Function to find maximum of three numbers
def maximum(a, b, c):
    return max(a, b, c)

print("Maximum number is:", maximum(10, 25, 15))
```

```python
# 2. Function to return distinct elements from a list
def distinct_list(lst):
    new_list = []
    
    for item in lst:
        if item not in new_list:
            new_list.append(item)
    
    return new_list

print("Distinct elements:", distinct_list([1, 2, 2, 3, 4, 4, 5]))

```

```python
# 3. Function to multiply all numbers in a list
def multiply_list(lst):
    result = 1
    
    for num in lst:
        result = result * num
    
    return result

print("Multiplication of list elements:", multiply_list([2, 3, 4]))
```

```python
# 4. Function to calculate factorial
def factorial(n):
    fact = 1
    
    for i in range(1, n + 1):
        fact = fact * i
    
    return fact

print("Factorial is:", factorial(5))
```

```python
# 5. Program to reverse a string
def reverse_string(text):
    return text[::-1]

print("Reversed string:", reverse_string("Python"))
```

```python
# 6. Function to check whether a number falls within a range
def check_range(num, start, end):
    if num in range(start, end + 1):
        print(num, "is within the range")
    else:
        print(num, "is outside the range")

check_range(7, 1, 10)
```

```python
# 7. Function to print even numbers from a given list
def even_numbers(lst):
    print("Even numbers are:")
    
    for num in lst:
        if num % 2 == 0:
            print(num)

even_numbers([1, 2, 3, 4, 5, 6, 7, 8])
```

```python
# 8. Function to check whether a number is prime
def prime_number(n):
    
    if n <= 1:
        return False
    
    for i in range(2, n):
        if n % i == 0:
            return False
    
    return True

num = 11

if prime_number(num):
    print(num, "is a Prime Number")
else:
    print(num, "is not a Prime Number")
```

```python
# 9. Function to count uppercase and lowercase letters
def count_case(text):
    upper = 0
    lower = 0
    
    for ch in text:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1
    
    print("Uppercase letters:", upper)
    print("Lowercase letters:", lower)

count_case("Hello Python")

```
