```python
import csv

# File name
file_name = "address_book.csv"

# Column headings
fields = ["Name", "Address", "Mobile", "Email"]

# Empty list to store records
records = []

# Taking input from user
n = int(input("How many records do you want to enter? "))

for i in range(n):
    print(f"\nEnter details for person {i+1}")

    name = input("Enter Name: ")
    address = input("Enter Address: ")
    mobile = input("Enter Mobile Number: ")
    email = input("Enter Email: ")

    records.append([name, address, mobile, email])

# Writing data into CSV file
with open(file_name, mode="w", newline="") as file:
    writer = csv.writer(file)

    # Writing column names
    writer.writerow(fields)

    # Writing records
    writer.writerows(records)

print("\nCSV file created successfully!")
```

```python
import sqlite3

conn = sqlite3.connect("college.db")

cur = conn.cursor()


# 1. Create Tables
cur.execute("""
CREATE TABLE IF NOT EXISTS Students (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    course TEXT
)
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS Teachers (
    teacher_id INTEGER PRIMARY KEY,
    name TEXT,
    subject TEXT
)
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS Courses (
    course_id INTEGER PRIMARY KEY,
    course_name TEXT,
    duration TEXT
)
""")

print("Tables created successfully!")


# 2. Insert Records
cur.execute("INSERT INTO Students VALUES (1, 'Rahul', 20, 'Python')")
cur.execute("INSERT INTO Students VALUES (2, 'Priya', 21, 'Java')")
cur.execute("INSERT INTO Students VALUES (3, 'Aman', 22, 'C++')")

cur.execute("INSERT INTO Teachers VALUES (1, 'Sharma Sir', 'Python')")
cur.execute("INSERT INTO Teachers VALUES (2, 'Mehta Sir', 'Java')")

cur.execute("INSERT INTO Courses VALUES (1, 'Python', '3 Months')")
cur.execute("INSERT INTO Courses VALUES (2, 'Java', '4 Months')")

conn.commit()

print("Records inserted successfully!")


# 3. Select Operations
print("\nAll Students:")
cur.execute("SELECT * FROM Students")
for row in cur.fetchall():
    print(row)

print("\nStudents Age Greater Than 20:")
cur.execute("SELECT * FROM Students WHERE age > 20")
for row in cur.fetchall():
    print(row)

print("\nOnly Student Names:")
cur.execute("SELECT name FROM Students")
for row in cur.fetchall():
    print(row)


# 4. Update Data
cur.execute("""
UPDATE Students
SET course = 'Data Science'
WHERE name = 'Rahul'
""")

conn.commit()

print("\nData Updated Successfully!")

# Check updated data
cur.execute("SELECT * FROM Students")
for row in cur.fetchall():
    print(row)


# 5. Delete Data
cur.execute("""
DELETE FROM Students
WHERE name = 'Aman'
""")

conn.commit()

print("\nData Deleted Successfully!")

# Final Table
print("\nFinal Student Table:")
cur.execute("SELECT * FROM Students")
for row in cur.fetchall():
    print(row)

# Close connection
conn.close()
```
