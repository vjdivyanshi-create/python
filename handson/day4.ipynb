```python
# if the no of arguments are not fixed then we can use * before the parameter name to accept variable number of arguments

def function(*city):
    print(city)

function("delhi", "mumbai", "kolkata", "chennai")
```

```python
# to specify that a fun can only take positional augments we put / sepreated with , after parameter name 
def function(a,/):
    print(a,"\n")

function(10)
# for only key word augemented we place * before parameter name
def function(*,a):
    print(a,"\n")    

function(a=10)
# to combine these two states ex = fun(a,b,/,*,c,d) where a,b are positional only and c,d are keyword only
def function(a,b,/,*,c,d):
    print(a)
    print(b)
    print(c)
    print(d)
function(10,20,c=30,d=40)
```

```python
# Recursion
def rec(k):
    if(k>=0):
        print(k)
        k = k-1
        rec(k)
rec(6)
```
