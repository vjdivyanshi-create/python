```python
import csv

d = [
    ["Name","Email","phone"],
    ["john doe ","john123@gmail.com ","1234567890"],
    ["rose ","rose_porter123@gmail.com ","0987654321"],
    ["michael ","michael_smith@gmail.com ","1122334455"],
    ["lisa ","lisa_jones@gmail.com ","2233445566"],
    ["emma ","emma_brown@gmail.com ","3344556677"]
]
with open('data.csv', 'w', newline='') as file:
    write = csv.writer(file)

    for row in d:
        write.writerow(row)
```

```python
# Error handling( type Error )
try:
    n1 = float(input("enter denominator:"))
    n2 = float(input("enter numerator:"))
    res = float(n2/n1)
    print(res)
except ZeroDivisionError:
    print("denominator cannot be zero")
except ValueError:
    print("invalid input")
```

```python
# Name Error 
# try and except is also used as if-else statement.
try:
    n1 = [1,2,3,4]
    print(n1[4])
except IndexError:
    print("the requested index is not found in the search")
```
