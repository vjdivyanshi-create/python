```python
# copy vs view
a = np.array([1,2,3,4,5,6,7,8])
x=a.copy()
x[0]=100
print(x)
print(a)# value of a is no changed

print("---------------")
y = a.view()
y[2]=800
print(y)
print(a) # value of a is changed

print("---------------")
print(a.reshape(4,2))
```
