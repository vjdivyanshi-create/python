```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df2 = pd.DataFrame({"A":[10,20,30],"B":[50,60,70]}, index=[1,2,3])
print(df1,"\n")
print(df2,"\n")

res = df1+df2
print(res)

res = df2-df1
print(res)
```

```python
one = pd.DataFrame({
    "Name":['A1','A2','A3','A4'],
    "Subject":['sub1','sub2','sub3','sub4'],
    "Marks":[73,92,68,84]
})

print(one,'\n')

two = pd.DataFrame({
    "Name":['B1','B2','B3','B4'],
    "Subject":['sub5','sub6','sub7','sub8'],
    "Marks":[83,62,78,94]
},index=[1,2,3,4])

print(two,"\n")

res = pd.concat([one,two], keys=["x","y"])
print(res)
```

```python
res = pd.concat([one,two],axis=1)
print(res,"\n")

res = pd.concat([one,two], ignore_index=True)
print(res)
```

```python
# Merging
one = pd.DataFrame({
    "id":[1,2,3,4],
    "Name":["a1","a2","a3","a4"],
    "sub":["eng","maths","phy","cpp"]
})

two= pd.DataFrame({
    "id":[2,1,3,4],
    "Name":["B1","b2","b3","b4"],
    "sub":["python","java","c++","c"],
    "marks":[20,19,18,15]
})

print(one,"\n")
print(two,"\n")

print("merged datafarme:")
res = one.merge(two, on="id")
print(res)
```

```python
one = pd.DataFrame({
    "id":[1,2,3,4],
    "Name":["a1","a2","a3","a4"],
    "sub":["python","maths","phy","java"]
})

two= pd.DataFrame({
    "id":[2,1,3,4],
    "Name":["B1","b2","b3","b4"],
    "sub":["python","java","c++","c"],
    "marks":[20,19,18,15]
})
res = one.merge(two ,on="sub",how="left")
print(res)
```

```python
one = pd.DataFrame({
    "id":[1,2,3,4],
    "Name":["a1","a2","a3","a4"],
    "sub":["python","maths","phy","java"]
})

two= pd.DataFrame({
    "id":[2,1,3,4],
    "Name":["B1","b2","b3","b4"],
    "sub":["python","java","c++","c"],
    "marks":[20,19,18,15]
})

res = one.join(two,lsuffix="left",rsuffix="right")
print(res)
```
