```python

name = input("Enter student name: ")
student_class = input("Enter class: ")

mark1 = float(input("Enter marks of Subject 1: "))
mark2 = float(input("Enter marks of Subject 2: "))
mark3 = float(input("Enter marks of Subject 3: "))
mark4 = float(input("Enter marks of Subject 4: "))
mark5 = float(input("Enter marks of Subject 5: "))

total = mark1 + mark2 + mark3 + mark4 + mark5
percentage = (total / 500) * 100

print("\n----- Student Result -----")
print("Name :", name)
print("Class :", student_class)
print("Total Marks :", total)
print("Percentage :", percentage, "%")
```

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

result = str1 + " " + str2

combined = result

print("\nConcatenated String:", combined)

# String methods
print("\n--- String Methods Output ---")

print("lower()      :", combined.lower())
print("upper()      :", combined.upper())
print("title()      :", combined.title())
print("swapcase()   :", combined.swapcase())
print("capitalize() :", combined.capitalize())
print("casefold()   :", combined.casefold())

print("center()     :", combined.center(50, "*"))

# count()
word = input("\nEnter word/character to count: ")
print("count()      :", combined.count(word))

# endswith()
end_word = input("Enter ending word to check: ")
print("endswith()   :", combined.endswith(end_word))

# find()
find_word = input("Enter word to find: ")
print("find()       :", combined.find(find_word))

# Check methods
print("isalnum()    :", combined.isalnum())
print("isdigit()    :", combined.isdigit())
print("isnumeric()  :", combined.isnumeric())
print("isspace()    :", combined.isspace())

# replace()
old_word = input("Enter word to replace: ")
new_word = input("Enter new word: ")
print("replace()    :", combined.replace(old_word, new_word))
```

```python
a = 10
print("Initial value of a =", a)

# += operator
a += 5
print("After a += 5 :", a)

# -= operator
a -= 3
print("After a -= 3 :", a)

# *= operator
a *= 2
print("After a *= 2 :", a)

# /= operator
a /= 4
print("After a /= 4 :", a)

# %= operator
a %= 3
print("After a %= 3 :", a)

# **= operator
a **= 2
print("After a **= 2 :", a)

# //= operator
a //= 2
print("After a //= 2 :", a)

# Bitwise assignment operators
b = 6
print("\nInitial value of b =", b)

# &= operator
b &= 3
print("After b &= 3 :", b)

# |= operator
b |= 5
print("After b |= 5 :", b)

# ^= operator
b ^= 2
print("After b ^= 2 :", b)

# <<= operator
b <<= 1
print("After b <<= 1 :", b)

# >>= operator
b >>= 2
print("After b >>= 2 :", b)
```

```python
name = input("Enter student name: ")
student_class = input("Enter class: ")

mark1 = float(input("Enter marks of Subject 1: "))
mark2 = float(input("Enter marks of Subject 2: "))
mark3 = float(input("Enter marks of Subject 3: "))
mark4 = float(input("Enter marks of Subject 4: "))
mark5 = float(input("Enter marks of Subject 5: "))

total = mark1 + mark2 + mark3 + mark4 + mark5
percentage = (total / 500) * 100

if percentage >= 60:
    grade = 'A'
elif percentage >= 50 and percentage < 60:
    grade = 'B'
elif percentage >= 40 and percentage < 50:
    grade = 'C'
elif percentage >= 33 and percentage < 40:
    grade = 'D'
else:
    grade = 'Fail'

print("\n----- Student Result -----")
print("Name        :", name)
print("Class       :", student_class)
print("Total Marks :", total)
print("Percentage  :", percentage, "%")
print("Grade       :", grade)
```
